## Question 2 :
### Given a DNA sequence and a state path sequence, find the probability that the given DNA sequence is generated from the given path sequence only.

####  Input DNA sequence: CTTCATGTGAAAGCAGACGTAAGTCA 
####  Input State path sequence: EEEEEEEEEEEEEEEEEE5IIIIIII$
####  Emission & Transition probabilities can be found from above figure
####  Output: State path probability for above state path [Log probability required i.e. log(p)]




```python
import math
sequence="CTTCATGTGAAAGCAGACGTAAGTCA"
state_path="EEEEEEEEEEEEEEEEEE5IIIIIII$"
prob=1
emmision={'E':{'A':0.25,'C':0.25,'G':0.25,'T':0.25},'5':{'A':0.05,'C':0,'G':0.95,'T':0},'I':{'A':0.4,'C':0.1,'G':0.1,'T':0.4}}
transition={'^':{'E':1},'E':{'E':0.9,'5':0.1},'5':{'I':1},'I':{'I':0.9,'$':0.1}}
for i in range(len(sequence)):
    s0=state_path[i]
    s1=state_path[i+1]
    n=sequence[i]
    e=emmision[s0][n]
    t=transition[s0][s1]
    prob=prob*e*t
print(prob)
print("State path probability (log) for above state path is : "+str(math.log(prob)))
```
